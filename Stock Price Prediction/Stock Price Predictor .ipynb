```python
import matplotlib.pyplot as plt
import seaborn as sns

def plot_graph(figsize, data, label, color='dodgerblue'):
    plt.figure(figsize=figsize)
    
    # Apply seaborn style for better visuals
    sns.set_style("whitegrid")
    
    # Plotting the data
    plt.plot(data, label=label, linewidth=2.5, color=color)

    # Styling the graph
    plt.title(f"{label} Over Time", fontsize=16, weight='bold', color='navy')
    plt.xlabel("Time", fontsize=12)
    plt.ylabel("Value", fontsize=12)
    
    plt.xticks(rotation=45)
    plt.legend(loc='upper left')
    
    plt.tight_layout()  # Adjust layout to prevent clipping
    plt.show()
```

```python
plot_graph((15, 5), google_data['MA_for_250_days'], 'MA_for_250_days')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the style
sns.set(style="whitegrid", palette="pastel", font_scale=1.2)

# Plotting function
def plot_graph(size, data, highlight_column):
    plt.figure(figsize=size)

    for column in data.columns:
        # Convert column name to string to avoid warning
        label_str = str(column)
        series = data[column]

        if column == highlight_column:
            plt.plot(series, label=label_str, linewidth=2.5, color='tomato')
        else:
            plt.plot(series, label=label_str, linewidth=1.5)

    plt.title("Stock Close Price vs 250-Day Moving Average", fontsize=18, weight='bold')
    plt.xlabel("Date", fontsize=14)
    plt.ylabel("Price", fontsize=14)
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.show()

# Example usage
plot_graph((15, 5), google_data[['Close', 'MA_for_250_days']], 'MA_for_250_days')
```

```python
Close_price=google_data[['Close']]
max(Close_price.values),min(Close_price.values)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(Close_price)
scaled_data

```

```python
len(scaled_data)
```

```python
x_data=[]
y_data=[]
for i in range(100,len(scaled_data)):
    x_data.append(scaled_data[i-100:i])
    y_data.append(scaled_data[i])
import numpy as np
x_data,y_data=np.array(x_data),np.array(y_data)
```

```python
x_data[0],y_data[0]
```

```python
int(len(x_data)*0.7)
```

```python
5032-int(len(x_data)*0.7)
```

```python
splitting_len=int(len(x_data)*0.7)
x_train=x_data[:splitting_len]
y_train=y_data[:splitting_len]

x_test=x_data[splitting_len:]
y_test=y_data[splitting_len:]
```

```python
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM, Input

model = Sequential()
model.add(Input(shape=(x_train.shape[1], 1)))  # Preferred over passing input_shape to LSTM
model.add(LSTM(128, return_sequences=True))
model.add(LSTM(64, return_sequences=False))
model.add(Dense(25))
model.add(Dense(1))

model.compile(optimizer='adam', loss='mean_squared_error')
```

```python
model.fit(x_train,y_train,batch_size=1,epochs=2)
```

```python
model.summary()
```

```python
predictions=model.predict(x_test)
```

```python
predictions
```

```python
inv_predictions=scaler.inverse_transform(predictions)
inv_predictions
```

```python
y_test
```

```python
inv_y_test=scaler.inverse_transform(y_test)
inv_y_test
```

```python
rmse=np.sqrt(np.mean((inv_predictions-inv_y_test)**2))
```

```python
rmse
```

```python
plotting_data=pd.DataFrame(
    {
        'original_test_data':inv_y_test.reshape(-1),
        'prediction':inv_predictions.reshape(-1)
    },
    index=google_data.index[splitting_len+100:]
)
plotting_data.head()
```

```python
plot_graph((15,6),plotting_data,'test data')
```

```python
plot_graph((15, 6), pd.concat([Close_price[:splitting_len+100], plotting_data], axis=0), 'Whole data')
```
